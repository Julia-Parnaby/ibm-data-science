```python
# Function to determine marker colour, based on class value
def set_marker_colour(class_value):
    if class_value == 1:
        return 'green'
    elif class_value == 0:
        return 'red'
    else:
        return 'unknown'  # Handle unexpected values

# Apply the function to the 'class' column and create the new 'marker_colour' column
spacex_df['marker_colour'] = spacex_df['class'].apply(set_marker_colour)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
